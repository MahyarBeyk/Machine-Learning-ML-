## K-Means

### a) Importing database

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.set()
```

```python
from sklearn.datasets import make_blobs

X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.40, random_state=0)
```

```python
plt.scatter(X[:,0], X[:,1], s=50)
plt.show()
```

### b) Model

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans = kmeans.fit(X)
```

### c) Evaluating model

```python
y_kmeans = kmeans.predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=y_kmeans, s=50, cmap='viridis')
```

```python
centers = kmeans.cluster_centers_
centers
```

```python
plt.scatter(centers[:,0], centers[:,1], c='black', s=200, alpha=0.5)
plt.show()
```

```python
plt.scatter(X[:,0], X[:,1], c=y_kmeans, s=50, cmap='viridis')

plt.scatter(centers[:,0], centers[:,1], c='black', s=200, alpha=0.5)
plt.show()
```

## MeanShift

### a) Importing database

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
style.use('ggplot')
```

```python
from sklearn.datasets import make_blobs

centers = [[2,2], [4,5], [3,10]]
X, _ = make_blobs(n_samples=500, centers=centers, cluster_std=1)
```

```python
plt.scatter(X[:,0], X[:,1])
plt.show()
```

### b) Model

```python
from sklearn.cluster import MeanShift
```

```python
ms = MeanShift()
```

```python
ms = ms.fit(X)
```

```python
labels = ms.labels_

labels
```

```python
cluster_centers = ms.cluster_centers_

cluster_centers
```

```python
n_clusters = len(np.unique(labels))

print('Estimated clusters: ', n_clusters)
```

```python
colors = 10*['r.', 'g.', 'b.', 'c.', 'k.', 'y.', 'm.']
```

```python
for i in range(len(X)):
    plt.plot(X[i][0], X[i][1], colors[labels[i]], markersize=10)
    plt.scatter(cluster_centers[:,0], cluster_centers[:,1], marker='x', color='k', s=150, linewidths=5, zorder=10)
    
plt.show()
```

## Validating with Silhouette

### a) Importing database

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
from sklearn.datasets import make_blobs

X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.40, random_state=0)
```

### b) Model

```python
from sklearn.cluster import KMeans
```

```python
from sklearn import metrics
```

```python
scores = []
```

```python
values = np.arange(2, 10)

for num_clusters in values:
    kmeans = KMeans(init='k-means++', n_clusters=num_clusters, n_init=10)
    kmeans = kmeans.fit(X)
```

```python
score = metrics.silhouette_score(X, kmeans.labels_, metric='euclidean', sample_size=len(X))
```

```python
print('\n Number of clusters = ', num_clusters)
print('Silhouette score = ', score)
scores.append(score)
```

```python
num_clusters = np.argmax(scores) + values[0]

print('\n Optimal number of clusters = ', num_clusters)
```

## KNN

### a) Importing database

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[3.1,2.3], [2.3,4.2], [3.9,3.5], [3.7,6.4], [4.8,1.9], [8.3,3.1], [5.2,7.5], [4.8,4.7], [3.5,5.1], [4.4,2.9]])

A
```

```python
k=3
```

```python
test_data = [3.3,2.9]
```

```python
plt.figure()
plt.title('Input data')
plt.scatter(A[:,0], A[:,1], marker='o', s=100, color='black')
plt.show()
```

### b) Model

```python
from sklearn.neighbors import NearestNeighbors
```

```python
knn_model = NearestNeighbors(n_neighbors=k, algorithm='auto')
```

```python
knn_model = knn_model.fit(A)
```

```python
distances, indices = knn_model.kneighbors([test_data])

distances, indices
```

```python
print('\n K Nearest Neighbors: ')
for rank, index in enumerate(indices[0][:k], start=1):
    print(str(rank)+ ' is ', A[index])
```

```python
plt.figure()
plt.title('Nearest neighbors')
plt.scatter(A[:,0], A[:,1], marker='o', s=100, color='k')
```

```python
plt.scatter(A[indices][0][:][:,0], A[indices][0][:][:,1], marker='o', s=250, color='k', facecolors='none')
```

```python
plt.scatter(test_data[0], test_data[1], marker='x', s=100, color='k')
```

```python
plt.figure()
plt.title('Nearest neighbors')
plt.scatter(A[:,0], A[:,1], marker='o', s=100, color='k')
plt.scatter(A[indices][0][:][:,0], A[indices][0][:][:,1], marker='o', s=250, color='k', facecolors='none')
plt.scatter(test_data[0], test_data[1], marker='x', s=100, color='k')
plt.show()
```
