## SVM

### a) Importing database

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris
```

### b) Splitting data

```python
#Making 2D data

x = iris.data[:, :2]
x
```

```python
y = iris.target

y
```

### c) Making meshgrid

```python
x_min, x_max = x[:, 0].min()-1, x[:,0].max()+1
y_min, y_max = x[:, 1].min()-1, x[:, 1].max()+1
mesh_step_size = (x_max - x_min)/100

x_min, y_min, x_max, y_max, mesh_step_size
```

```python
xx, yy = np.meshgrid(np.arange(x_min, x_max, mesh_step_size), np.arange(y_min, y_max, mesh_step_size))

xx, yy
```

```python
# ravel() makes 2D points from xx and yy lists

x_plot = np.c_[xx.ravel(), yy.ravel()]

x_plot
```

### d) Model

```python
from sklearn import svm
```

```python
# Regularization parameter

c = 1.0
```

```python
svc_classifier = svm.SVC(kernel='linear', C=c, decision_function_shape='ovr')

svc_classifier
```

```python
svc_classifier = svc_classifier.fit(x,y)
```

### e) Evaluating model

```python
z = svc_classifier.predict(x_plot)
```

```python
z = z.reshape(xx.shape)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.contourf(xx, yy, z, cmap=plt.cm.tab10, alpha=0.3)
plt.scatter(x[:,0], x[:,1], c=y, cmap=plt.cm.Set1)

plt.xlabel('Sepal length')
plt.ylabel('Sepal with')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with linear kernel')
plt.show()
```

## Logistic regression

### a) Importing database

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[2,4.8], [2.9,4.7], [2.5, 5], [3.2,5.5], [6,5], [7.6,4], [3.2,0.9], [2.9,1.9], [2.4,3.5], [0.5,3.4], [1,4], [0.9,5.9]])
y = np.array([0,0,0,1,1,1,2,2,2,3,3,3])

x, y
```

### b) Model

```python
from sklearn import linear_model
```

```python
classifier_LR = linear_model.LogisticRegression(solver='liblinear', C=75)
```

```python
classifier_LR = classifier_LR.fit(x,y)
```

```python
def logistic_visualize(classifier_LR, x, y):
    x_min, x_max = x[:, 0].min()-1, x[:,0].max()+1
    y_min, y_max = x[:, 1].min()-1, x[:, 1].max()+1
    mesh_step_size = (x_max - x_min)/100
    return x_min, y_min, x_max, y_max, mesh_step_size
```

```python
x_min, y_min, x_max, y_max, mesh_step_size = logistic_visualize(classifier_LR,x,y)
```

```python
x_vals, y_vals = np.meshgrid(np.arange(x_min, x_max, mesh_step_size), np.arange(y_min, y_max, mesh_step_size))

x_vals, y_vals
```

```python
x_plot = np.c_[x_vals.ravel(), y_vals.ravel()]

x_plot
```

### c) Evaluating model

```python
output = classifier_LR.predict(np.c_[x_vals.ravel(), y_vals.ravel()])
```

```python
output = output.reshape(x_vals.shape)
```

```python
plt.figure()
plt.pcolormesh(x_vals, y_vals, output, cmap=plt.cm.gray)
plt.scatter(x[:,0], x[:,1], c=y, s=75, edgecolors='black', linewidth=1, cmap=plt.cm.Paired)

plt.xlim(x_vals.min(), x_vals.max())
plt.ylim(y_vals.min(), y_vals.max())
plt.xticks((np.arange(int(x[:,0].min()-1), int(x[:,0].max()+1),1.0)))
plt.yticks((np.arange(int(x[:,1].min()-1), int(x[:,1].max()+1),1.0)))
plt.show()

#logistic_visualize(classifier_LR,x,y)
```

## Decision Tree

### a) Importing database

```python
import pydotplus
from sklearn import tree
```

```python
from sklearn.datasets import load_iris
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
import collections
```

```python
x = [[165,19], [175,32], [136,35], [174,65], [141,28], [176,15], [131,32], [166,6], [128,32], [179,10], [136,34], [186,2], [126,25], [176,28], [112,38], [169,9], [171,36], [116,25], [196,25]]
y = ['man', 'woman', 'woman', 'man', 'woman', 'man', 'woman', 'man', 'woman', 'man', 'woman', 'man', 'woman', 'woman', 'woman', 'man', 'woman', 'woman', 'man']
data_feature_names = ['heigh', 'length of hair']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.40, random_state=5)
```

### b) Model

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf = clf.fit(x_train,y_train)
```

### c) Evaluating model

```python
prediction = clf.predict([[133,37]])
print(prediction)
```

```python
preds = clf.predict(x_test, y_test)
preds
```

```python
dot_data = tree.export_graphviz(clf,feature_names=data_feature_names, out_file=None, filled=True, rounded=True)
graph = pydotplus.graph_from_dot_data(dot_data)
colors = ('orange', 'yellow')
edges = collections.defaultdict(list)
for edge in graph.get_edge_list():
    edges[edge.get_source()].append(int(edge.get_destination()))
for edge in edges:
    edges[edge].sort()
for i in range(2):
    dest = graph.get_node(str(edges[edge][i]))[0]
    dest.set_fillcolor(colors[i])
graph.write_png('DecisionTree.png')
```

## Random forest

### a) Importing database

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
cancer = load_breast_cancer()
```

```python
x_train, x_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)
```

### b) Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest = RandomForestClassifier(n_estimators=50, random_state=0)
```

```python
forest = forest.fit(x_train, y_train)
```

```python
Accuracy_train = forest.score(x_train, y_train)
Accuracy_test = forest.score(x_test, y_test)

print("Accuracy on the training subset: (:.3f)",format(Accuracy_train))
print("Accuracy on the test subset: (:.3f)",format(Accuracy_test))
```

```python
n_features = cancer.data.shape[1]

n_features
```

```python
plt.barh(range(n_features), forest.feature_importances_, align='center')

plt.yticks(np.arange(n_features), cancer.feature_names)
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.show()
```

# Supervised learning: 2) Regression

## Linear Regression

### a) Importing database

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.metrics as sm
```

```python
x = 10* np.random.normal(0,1,10)
y = 10*np.random.normal(0,1,10)

x,y
```

### b) Model

```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
```

```python
reg = reg.fit(x.reshape(-1,1), y.reshape(-1,1))
```

### c) Evaluating model

```python
y_test_pred = reg.predict(x.reshape(-1,1))
```

```python
plt.scatter(x, y, color='red')
plt.plot(x, y_test_pred, color='black', linewidth=2)

plt.xticks(())
plt.yticks(())
plt.show()
```

```python
Accuracy_mean_absolute_error = round(sm.mean_absolute_error(x,y),2)
Accuracy_mean_squared_error = round(sm.mean_squared_error(x,y),2)
Accuracy_median_absolute_error = round(sm.median_absolute_error(x,y),2)
Accuracy_explained_variance_score = round(sm.explained_variance_score(x,y),2)
Accuracy_r2_score = round(sm.r2_score(x,y),2)

print("Performance of Linear regression:")
print("Mean absolute error", Accuracy_mean_absolute_error)
print("Mean squared error", Accuracy_mean_squared_error)
print("Median absolute error", Accuracy_median_absolute_error)
print("Explained variance score", Accuracy_explained_variance_score)
print("R2 score", Accuracy_r2_score)
```

## Polynomial regression

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
import sklearn.metrics as sm
```

```python
x = 10* np.random.normal(0,1,50)
y = 10*np.random.normal(0,1,50)

x,y
```

```python
lm = LinearRegression()
```

```python
lm = lm.fit(x.reshape(-1,1), y.reshape(-1,1))
```

```python
y_pred = lm.predict(x.reshape(-1,1))

y_pred
```

```python
Input = [('polynomial', PolynomialFeatures(degree=10)),('modal', LinearRegression())]
pipe = Pipeline(Input)
pipe.fit(x.reshape(-1,1), y.reshape(-1,1))
poly_pred = pipe.predict(x.reshape(-1,1))
sorted_zip = sorted(zip(x, poly_pred))
x_poly, poly_pred = zip(*sorted_zip)

print("Linear Regression's Accuracy: \n", np.sqrt(mean_squared_error(y, y_pred)),'\n')
print("Polynomial Regression's Accuracy: \n", np.sqrt(mean_squared_error(y, poly_pred)))
```

```python
plt.scatter(x, y, color='red')
plt.plot(x, poly_pred, color='black', linewidth=2)

plt.xticks(())
plt.yticks(())
plt.show()
```
